```python
%matplotlib notebook

import pandas as pd
from matplotlib import pyplot as plt
```

```python
filepath1 = './models/chollet_balanced/chollet_bal_acc.csv'
filepath2 = './models/chollet_imbalanced/chollet_imbal_acc.csv'
filepath3 = './models/simple_balanced/simple_bal_acc.csv'
filepath4 = './models/simple_imbalanced/simple_imbal_acc.csv'
filepath5 = './models/simple_balanced/simple_bal_acc50.csv'

file1 = pd.read_csv(filepath1)
file2 = pd.read_csv(filepath2)
file3 = pd.read_csv(filepath3)
file4 = pd.read_csv(filepath4)
file5 = pd.read_csv(filepath5)

c_bal_acc = pd.DataFrame(file1)
c_imbal_acc = pd.DataFrame(file2)
s_bal_acc = pd.DataFrame(file3)
s_imbal_acc = pd.DataFrame(file4)
s_bal50 = pd.DataFrame(file5)
```

```python
#Chollet CNN balanced accuracy. #Chollet is the last name of whose model we based ours on
x1 = c_bal_acc.index
y1 = c_bal_acc['val_accuracy']
y2 = c_bal_acc['accuracy']

plt.plot(x1, y1, label='val accuracy')
plt.plot(x1, y2, label='accuracy')

plt.title('Training CNN')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()

plt.show()
plt.savefig('Chollet_balanced.png')
```

```python
#Chollet CNN imbalanced data

x2 = c_imbal_acc.index
y3 = c_imbal_acc['val_accuracy']
y4 = c_imbal_acc['accuracy']

plt.plot(x2, y3, label='val accuracy')
plt.plot(x2, y4, label='accuracy')

plt.title('Training CNN')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()

plt.savefig('chollet_imbalanced.png')
```

```python
#simple CNN model, balanced data

a1 = s_bal_acc.index
b1 = s_bal_acc['val_accuracy']
b2 = s_bal_acc['accuracy']

plt.plot(a1, b1, label='val accuracy')
plt.plot(a1, b2, label='accuracy')

plt.title('Training CNN')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()

plt.savefig('simple_balanced.png')
```

```python
#simple CNN model, imbalanced data

a2 = s_imbal_acc.index
b3 = s_imbal_acc['val_accuracy']
b4 = s_imbal_acc['accuracy']

plt.title('Training CNN')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()

plt.plot(a2, b3)
plt.plot(a2, b4)
```

```python
#simple model, balanced data across 50 epochs

a3 = s_bal50.index
b4 = s_bal50['val_accuracy']
b5 = s_bal50['accuracy']

plt.plot(a3, b4, label='val accuracy')
plt.plot(a3, b5, label='accuracy')

plt.title('Training CNN')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()

plt.savefig('simple_balanced_50.png')
```
